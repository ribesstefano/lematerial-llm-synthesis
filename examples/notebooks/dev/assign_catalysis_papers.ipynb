```python
import random

from datasets import load_dataset

dataset = load_dataset(
    "amayuelas/LeMat-Synth-Papers-Catalysis-v2", split="train"
)

dataset
```

```python
names = ["Sam", "Ana", "Alex N", "Faris", "Mathilde", "Ilya", "Magda"]

# sample 30 * len(names) random papers from the dataset


sampled_papers = random.sample(dataset["pdf_url"], 30 * len(names))
sampled_papers
```

```python
# for name in names, sample 30 random papers and assign to name, print to a .txt file with the pdf urls

for name in names:
    papers_for_name = random.sample(sampled_papers, 30)
    with open(f"{name}_papers.txt", "w") as f:
        for paper in papers_for_name:
            f.write(paper + "\n")
```
