```python
import random

import pandas as pd
from datasets import load_dataset

dataset = load_dataset(
    "LeMaterial/LeMat-Synth-Papers",
    name="full",
    split="thermocatalysis_keywords_and_LLM",
)

dataset
```

```python
df = dataset.to_pandas()
df.head()
```

```python
# count all unique vals in df['source']
df["source"].value_counts()
```

```python
names = ["Sam", "Ana", "Alex N", "Faris", "Mathilde", "Ilya", "Magda"]

# sample 30 * len(names) random papers from the dataset


sampled_papers = random.sample(dataset["pdf_url"], 30 * len(names))
sampled_papers
```

```python
# for name in names, sample 30 random papers and assign to name, print to a .txt file with the pdf urls

for name in names:
    papers_for_name = random.sample(sampled_papers, 30)
    with open(f"{name}_papers.txt", "w") as f:
        for paper in papers_for_name:
            f.write(paper + "\n")
```

```python
dataset_kw = load_dataset(
    "LeMaterial/LeMat-Synth-Papers",
    name="full",
    split="thermocatalysis_keywords_only",
)

df_kw = dataset_kw.to_pandas()
df_kw["source"].value_counts()
```

```python
# for df['source'].value_counts() and df_kw['source'].value_counts(), compare the two and see how many % of papers come from each source in both datasets
df_source_counts = df["source"].value_counts(normalize=True)
df_kw_source_counts = df_kw["source"].value_counts(normalize=True)

comparison = pd.DataFrame(
    {"LLM_and_Keywords": df_source_counts, "Keywords_Only": df_kw_source_counts}
).fillna(0)
comparison
```
